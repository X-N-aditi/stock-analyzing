# Apple Stock Analyzing

```python
# import library

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load the dataset

apple_stock = pd.read_csv("C:/Users/hp/Downloads/archive (4)/apple_stock.csv")
```

```python
# Check Column information of Dataset

apple_stock.info()
```

```python
# Converting Date column in DateTime Format and setting is as Index

apple_stock['Date'] = pd.to_datetime(apple_stock['Date'])
```

```python
# After converting checking whole dataset

apple_stock.head()
```

```python
# High column make a selected_column for analyzing its average

selected_column = apple_stock['High']

if 'High' in apple_stock.columns:
    average_value = selected_column.mean()
    print(f"Average of 'High': {average_value}")
```

```python
# Ploting Bar for selected_column's all value

column_name = selected_column.name

plt.figure(figsize = (35, 10))
apple_stock[column_name].value_counts().plot.bar()
```

```python
# Plotting scatter Plot for Date and High column for visulaizing 

plt.scatter(apple_stock['Date'], apple_stock['High'], alpha=0.5, c='red')
```

```python
# Plotting heatmap for finding correaltion between all columns of the Dataset

sns.heatmap(apple_stock.corr(), annot=True, cmap='coolwarm', fmt=".2")
```

```python
# Plotting lineplot with marker o to visualize data between date and close price 

plt.figure(figsize=(12, 6))
sns.lineplot(data=apple_stock, x = 'Date', y = 'Close', label = 'Close Price History', marker = 'o', color='blue')
```

```python
# Insights of Data

print(f"Average value of '{selected_column}' : {average_value}.")
print("------------------------------------------------------------------------------------------------------------")
print(f"Correlation : {apple_stock['Adj Close'].corr(apple_stock['Close'])}")
```
